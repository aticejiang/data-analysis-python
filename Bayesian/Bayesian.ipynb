```python
def loadCsv(filename): #读取数据
    lines = csv.reader(open(filename, "r"))
    dataset = np.array(list(lines), dtype=float)
    return dataset

#测试
filename = 'pima-indians-diabetes.data.csv'
dataset = loadCsv(filename)
dataset
```

```python
dataLabels = ['1', '2', '3', '4', '5', '6', '7', '8', 'category']  #特征标签
data_list = [] #保存数据的临时列表
data_dict = {} #保存数据的字典，用于生成pandas数据结构DataFrame
for each_label in dataLabels: #提取信息，生成字典
    for each in dataset:
        data_list.append(each[dataLabels.index(each_label)])
    data_dict[each_label] = data_list
    data_list = []
data_pd = pd.DataFrame(data_dict) #生成pandas.DataFrame
data_pd #打印pandas.DataFrame
```

```python
from sklearn.model_selection import train_test_split
pd_train,pd_test = train_test_split(data_pd, test_size=0.2) #划分数据集为训练集和测试机
print(pd_train.shape)
pd_train
```

```python
class GaussianNB(object):
    
    def __init__(self, pd_train, pd_test):
        self.pd_train = pd_train
        self.pd_test = pd_test
        self.G_y = pd_train.groupby('category')
        self.mean = G_y.agg('mean')
        self.std = G_y.agg('std')
        self.feature_num = len(self.mean.iloc[0]) #特征数量
        self.total_category = len(G_y) #类别总数
    
    def conditional_probability(self, x):  #计算条件概率
        '''
        Parameters:
            x: 输入特征
        Returns:
            c_prob: y类条件概率的列表
        '''
        p_y = pd_train.value_counts(pd_train['category']) / len(pd_train['category'])
        c_prob = []
        for i in range(self.total_category):
            p = p_y.loc[float(i)]
            for j in range(self.feature_num):
                p *= norm.pdf(x[j], self.mean.iloc[i, j], self.std.iloc[i, j])
            c_prob.append(p)
        return c_prob
    
    def predict(self, x): #根据条件概率预测
        '''
        Parameters:
            x: 输入特征
        Returns:
            result: 概率最大值是哪一类别
        '''
        c_prob = self.conditional_probability(x)
        result = np.argmax(c_prob)
        return result
    
    def test_nb(self): #测试样本，返回预测值
        '''
        Parameters:
            self.pd_test: 包含类别的测试集
        Returns:
            result: 所有测试样本的预测值
        '''
        result = []
        for i in range(len(pd_test)):
            feature_tmp = self.pd_test.iloc[i].to_list()[:-1]
            r_tmp = self.predict(feature_tmp)
            result.append(r_tmp)
        return result
    
    def acc_nb(self): #测试样本准确率
        '''
        Parameters:
            self.pd_test: 包含类别的测试集
        Returns:
            acc: 测试集准确率
        '''
        result = self.test_nb()
        count = 0
        for i in range(len(result)):
            if result[i] == self.pd_test.iloc[:,-1].to_list()[i]:
                count += 1
        acc = count / len(result)
        return acc
```

```python
nb = GaussianNB(pd_train, pd_test)
nb.acc_nb()
```

使用sklearn的高斯朴素贝叶斯方法

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
model = GaussianNB()

model.fit(pd_train.iloc[:,:-1], pd_train.iloc[:,-1]) #训练模型
```

```python
predict_results=model.predict(pd_test.iloc[:,:-1]) #预测测试集
print(predict_results)
print(accuracy_score(predict_results, pd_test.iloc[:,-1].to_list())) #得到准确率
```
